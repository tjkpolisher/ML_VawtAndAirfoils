```python
import os
import sys
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
x = [16, 32, 64, 128, 256, 512]
l2_train_step1 = [0.4383, 0.2557, 0.1319, 0.0766, 0.0486, 0.0661]
#l2_train_step2 = []
```

```python
l2_test_step1 = [0.6525, 0.5523, 0.5410, 0.4956, 0.4963, 0.5033]
#l2_test_step2 = []
```

```python
plt.plot(x, l2_train_step1, 'b')
plt.xlabel('Number of units', fontsize=15)
plt.xscale("log", basex=2)
plt.ylabel('$L_2$ error norm', fontsize=15)
plt.title('$L_2$ error norm comparison (training, both layers)', fontsize=20)
plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.plot(x, l2_test_step1, 'b')
plt.xlabel('Number of units', fontsize=15)
plt.xscale("log", basex=2)
plt.ylabel('$L_2$ error norm', fontsize=15)
plt.title('$L_2$ error norm comparison (test, both layers)', fontsize=20)
plt.tight_layout()
plt.grid()
plt.show()
```
