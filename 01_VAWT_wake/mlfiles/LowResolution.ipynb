```python
TSR20_field.shape
```

```python
TSR20_field = TSR20_field.reshape(37, 100, 100, 1)
```

```python
input_img = tf.keras.Input(shape=(100, 100, 1), name='input_img')

x = tf.keras.layers.AveragePooling2D((2, 2), padding='same')(input_img)
x = tf.keras.layers.MaxPooling2D((2, 2), padding='same')(x)
decoded = tf.keras.layers.AveragePooling2D((2, 2))(x)

# 인코더 = 저해상도 이미지 구축
encoder = tf.keras.Model(input_img, decoded)
```

```python
encoder.summary()
```

```python
encoder.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-04),
                loss = tf.keras.losses.MeanSquaredError(),
                metrics=[tf.keras.metrics.RootMeanSquaredError(name="rmse")])
```

```python
encoded = encoder.predict(TSR20_field)
```
