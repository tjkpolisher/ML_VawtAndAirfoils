```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import datetime
%matplotlib inline
```

```python
x_train = np.linspace(0, 10, 11).reshape((-1,1))
y_train = x_train**2 + 2*x_train
x_test = np.array([0.3, 1.2, 6.5, 4.2, 3.9, 9.2, 2.1, 7.7, 8.5, 5.3]).reshape((-1,1))
y_test = x_test**2 + 2*x_test
```

```python
# Numbers of trees modification
for i in (10, 20, 30, 40, 50, 60):
    rfr = RandomForestRegressor(n_estimators = i, max_depth = 10, random_state = 15)
    rfr.fit(x_train, y_train)

    relation_square_train = rfr.score(x_train, y_train)
    print('결정계수(training)(%d): ' %i, relation_square_train)
    relation_square_test = rfr.score(x_test, y_test)
    print('결정계수(test)(%d): ' %i, relation_square_test)
    y_p_train = rfr.predict(x_train).reshape(-1, 1)
    y_p_test = rfr.predict(x_test).reshape(-1, 1)
    
    abs_error_train = y_p_train - y_train
    print(abs_error_train.T)
    l2_error_train = np.sqrt(np.sum((y_p_train - y_train)**2) / np.sum(y_train**2))
    print(l2_error_train)
    l2_error_train_list = []
    for j in range(0, x_train.shape[0]):
        l2_error_train_data = np.sqrt((y_p_train[j] - y_train[j])**2 / y_train[j]**2)
        l2_error_train_list.append(l2_error_train_data)
    print(l2_error_train_list)
    
    abs_error_test = y_p_test - y_test
    print(abs_error_test.T)
    l2_error_test = np.sqrt(np.sum((y_p_test - y_test)**2) / np.sum(y_test**2))
    print(l2_error_test)
    l2_error_test_list = []
    for j in range(0, x_test.shape[0]):
        l2_error_test_data = np.sqrt((y_p_test[j] - y_test[j])**2 / y_test[j]**2)
        l2_error_test_list.append(l2_error_test_data)
    print(l2_error_test_list)
    
    plt.scatter(x_train, y_train, marker = 'o', c='b', label='True value')
    plt.scatter(x_train, y_p_train, marker = 'x', c='r', label='Predicted value')
    plt.title('Training(n_estimators = %d)' %i, fontsize=15)
    plt.legend(loc='upper left')
    plt.grid()
    plt.tight_layout()
    
    plt.show()
    
    plt.scatter(x_test, y_test, marker = 'o', c='b', label='True value')
    plt.scatter(x_test, y_p_test, marker = 'x', c='r', label='Predicted value')
    plt.title('Test(n_estimators = %d)' %i, fontsize=15)
    plt.legend(loc='upper left')
    plt.grid()
    plt.tight_layout()
    
    plt.show()
```
